```python
converter = O.DataToNpyFiles(image_dir=aero_img_root,
                             mask_dir=aero_mask_root,
                             output_dir=output_dir,
                             images_file_name=images_name,
                             masks_file_name=masks_name,
                             image_extension='jpg',
                             mask_extension='png')
```
