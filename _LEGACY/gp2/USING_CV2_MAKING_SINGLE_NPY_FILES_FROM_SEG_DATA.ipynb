```python
import os
import cv2
import numpy as np

# Define the folder path
folder_path = gt_path

# Get the list of image file names in the folder
file_names = os.listdir(folder_path)

# Sort the file names alphabetically to ensure consistent ordering
file_names.sort()

# Initialize an empty list to store the image data
images = []

# Loop through each file in the folder
for file_name in file_names:
    # Read the image file
    file_path = os.path.join(folder_path, file_name)
    image = cv2.imread(file_path)

    # Convert the image to a NumPy array and append to the list
    images.append(image)

# Convert the list of images to a NumPy array
images_array = np.array(images)

# Save the array to a file
np.save(npy_img_path, images_array)
```
